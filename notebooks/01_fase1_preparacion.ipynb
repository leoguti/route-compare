## Fuentes de datos

Para este análisis utilizaremos las siguientes fuentes:

1. **Listado Maestro Trujillo v2**: Tabla maestra previa con información consolidada de rutas
2. **Datos Trujillo 2024**: Conjunto de datos actualizado con rutas operativas en 2024
3. **Carpeta kml-por-ruta**: Archivos KML individuales proporcionados por la autoridad de transporte
4. **Plan Regulador**: Documento oficial con rutas autorizadas y códigos oficiales
5. **OpenStreetMap**: Datos de rutas extraídos de OSM para la región de Trujillo
